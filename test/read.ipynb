```python
import os
import re

def extract_numbers_from_filenames(directory):
    # 检查目录是否存在
    if not os.path.isdir(directory):
        print(f"目录 {directory} 不存在。")
        return []
    
    # 获取目录中的所有文件和子目录
    filenames = os.listdir(directory)
    
    numbers = []
    for filename in filenames:
        # 提取所有包含小数点的数字序列
        numeric_parts = re.findall(r'-?\d+(?:\.\d+)?', filename)
        if numeric_parts:
            # 将每个数字序列转换为浮点数，并添加到numbers列表中
            for num_str in numeric_parts:
                number = float(num_str)
                numbers.append(number)
    
    return numbers

# 定义目录路径
directory_path = os.path.join('..', 'data', 'temp')

# 提取数字
extracted_numbers = extract_numbers_from_filenames(directory_path)

# 输出提取的数字
print("提取的数字有：")
print(extracted_numbers)
```

```python
from matplotlib import pyplot as plt
import numpy as np
# np.mean(extracted_numbers)
radii_median = np.median(extracted_numbers)
threshold = 0.8
filtered_data = np.array(extracted_numbers)[np.abs(np.array(extracted_numbers) - radii_median) <= threshold]
plt.boxplot(extracted_numbers, vert=False, patch_artist=True)
plt.xlabel('$R$')
plt.title('Boxplot of R Measurement')
plt.show()
```
